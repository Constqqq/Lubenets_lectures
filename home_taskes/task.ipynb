## Все классы в одном блоке

```python
class MixinSpeakable:
    def speak(self):
        return f"{self.name} говорит: \"{self.catchphrase}\""

class MixinActionable:
    def perform_action(self):
        return f"{self.name} делает: {self.signature_move}"
    def stomp_foot(self):
        return f"{self.name} топает ногой!"

class MixinCollectible:
    is_collectible = True
    def collect(self):
        return f"Фигурка {self.name} попала в коллекцию!"
    def display(self):
        return f"{self.name} теперь стоит на полке и радует глаз."

class MixinPoseable:
    def pose(self):
        return f"{self.name} принимает эффектную позу."
    def strike_pose(self):
        return f"{self.name} выделывается перед публикой."

class MixinCostumeWearable:
    def wear_costume(self):
        return f"На {self.name} надет костюм."
    def take_photo(self):
        return f"Сделано фото {self.name} в костюме."

class MixinAnimated:
    is_animated = True
    def animate(self):
        return f"{self.name} оживает на экране!"

class MixinFunny:
    def make_laugh(self):
        return f"{self.name} шутит: {self.joke}"

class BaseHuman(MixinSpeakable):
    is_human = True
    def breathe(self):
        return "Человек делает глубокий вдох."

class BaseCharacter(MixinSpeakable, MixinAnimated, MixinFunny):
    pass

class BaseInActionCharacter(BaseCharacter, MixinActionable):
    action_mode = True
    signature_move = "резко атакует"

class BaseFunkoPop(BaseCharacter, MixinCollectible, MixinPoseable):
    is_vinyl_figure = True

class BaseCosplayer(BaseCharacter, BaseHuman, MixinCostumeWearable, MixinPoseable):
    costume_quality = "высококачественный"
    props_count = 5


class Shrek(BaseCharacter):
    name = "Гринч"
    catchphrase = "Я люблю свои зелёные горы!"
    joke = "Мой друг слон обожает шутки о бананах."

class PussInBoots(BaseCharacter):
    name = "Кот-Ловкач"
    catchphrase = "Я прыгаю и всех удивляю!"
    joke = "Мои лапки слишком быстрые для своих штанов."

class Donkey(BaseCharacter):
    name = "Болтун"
    catchphrase = "Я могу говорить обо всём и сразу!"
    joke = "А вы знали, что я люблю морковь больше всех?"

class JackHorner(BaseCharacter):
    name = "Джек-Повар"
    is_villain = True
    catchphrase = "Я мастерски плету коварные планы."
    joke = "Иногда моя еда оживает и убегает!"


class ShrekInAction(Shrek, BaseInActionCharacter):
    quest = "Спасти деревню от нашествия троллей"
    signature_move = "Топает ногами и кричит!"

class PussInBootsInAction(PussInBoots, BaseInActionCharacter):
    quest = "Помогать друзьям в опасных миссиях"
    signature_move = "Быстро крутится и атакует!"

class DonkeyInAction(Donkey, BaseInActionCharacter):
    quest = "Поднимать настроение команде"
    signature_move = "Отвлекает врагов болтовнёй"

class JackHornerInAction(JackHorner, BaseInActionCharacter):
    quest = "Украсть самые вкусные пироги"
    signature_move = "Ловко бросает ингредиенты и смешит всех"


class ShrekFunkoPop(Shrek, BaseFunkoPop):
    rarity = "Супер редкий"
    pose_style = "стоя с поднятой лапой"

class PussInBootsFunkoPop(PussInBoots, BaseFunkoPop):
    rarity = "Легендарный"
    pose_style = "стоя с поднятым мечом"

class DonkeyFunkoPop(Donkey, BaseFunkoPop):
    rarity = "Редкий"
    pose_style = "стоя с открытым ртом"

class JackHornerFunkoPop(JackHorner, BaseFunkoPop):
    rarity = "Редкий"
    pose_style = "стоя с поднятой рукой"


class ShrekCosplayer(Shrek, BaseCosplayer):
    human_name = "имя человековое"
    props_count = 3
    makeup_required = True
    green_bodysuit = True

class PussInBootsCosplayer(PussInBoots, BaseCosplayer):
    human_name = "не имя не человековое"
    props_count = 4
    requires_sword = True
    hat_with_feather = True

class DonkeyCosplayer(Donkey, BaseCosplayer):
    human_name = "апщрцкщцкп"
    props_count = 2
    donkey_ears = True
    tail_attached = True

class JackHornerCosplayer(JackHorner, BaseCosplayer):
    human_name = "бамбамбам"
    props_count = 5
    requires_magical_props = True
    bag_of_artifacts = True

print("это фейк чтобы набрать class`ы")
```
