```python

import matplotlib.pyplot as plt

# Ваши x-значения (годы)
dob_array = np.array([1990, 1991, 1992, 1993, 1994, 1995, 1996, 1997, 1998, 1999, 2000])

# Генерация y-значений как индексы массива (0, 1, ..., N-1)
y_values = np.arange(len(dob_array))

# Рассчитайте средние значения
x_mean = np.mean(dob_array)
y_mean = np.mean(y_values)

# Ручной расчёт коэффициентов линейной регрессии
numerator = np.sum((dob_array - x_mean) * (y_values - y_mean))
denominator = np.sum((dob_array - x_mean) ** 2)
slope = numerator / denominator
intercept = y_mean - slope * x_mean

# Постройте график
plt.figure(figsize=(10, 6))
plt.scatter(dob_array, y_values, color='blue', label='Данные')  # Точки
plt.plot(
    dob_array,
    slope * dob_array + intercept,
    color='red',
    label=f'Тренд: y={slope:.2f}x+{intercept:.2f}'  # Линия регрессии
)

# Настройки графика
plt.title('Диаграмма рассеяния с трендовой линией')
plt.xlabel('Годы')
plt.ylabel('Номера данных в массиве')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Ваши x-значения (годы)
dob_array = np.array([1990, 1991, 1992, 1993, 1994, 1995, 1996, 1997, 1998, 1999, 2000])

# Y-значения как индексы (номера позиций в массиве)
y_values = np.arange(len(dob_array))

# Стиль сетки
sns.set(style="whitegrid")

# Построение графика
plt.figure(figsize=(10, 6))
sns.scatterplot(x=dob_array, y=y_values, color='blue', label='Данные')  # Точки
sns.regplot(x=dob_array, y=y_values, ci=None, color='red', label='Трендовая линия')  # Линия регрессии

# Настройки графика
plt.title('Диаграмма рассеяния с трендовой линией')
plt.xlabel('Годы')
plt.ylabel('Номера данных в массиве')
plt.legend()
plt.grid(True)
plt.show()
```

```python


# Ваши x-значения (годы)
dob_array = np.array([1990, 1991, 1992, 1993, 1994, 1995, 1996, 1997, 1998, 1999, 2000])

# Y-значения как индексы (номера позиций в массиве)
y_values = np.arange(len(dob_array))

# Создание словаря с данными
data = {
    "Годы": dob_array,
    "Номера данных": y_values
}

# Построение графика с Plotly Express
fig = px.scatter(
    data,
    x="Годы",
    y="Номера данных",
    trendline="ols",  # Добавление трендовой линии
    title="Диаграмма рассеяния с трендовой линией"
)

fig.show(renderer='browser')
```
