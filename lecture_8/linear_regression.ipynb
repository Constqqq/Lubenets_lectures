```python

import matplotlib.pyplot as plt

#  x- ()
dob_array = np.array([1990, 1991, 1992, 1993, 1994, 1995, 1996, 1997, 1998, 1999, 2000])

#  y-    (0, 1, ..., N-1)
y_values = np.arange(len(dob_array))

#   
x_mean = np.mean(dob_array)
y_mean = np.mean(y_values)

#     
numerator = np.sum((dob_array - x_mean) * (y_values - y_mean))
denominator = np.sum((dob_array - x_mean) ** 2)
slope = numerator / denominator
intercept = y_mean - slope * x_mean

#  
plt.figure(figsize=(10, 6))
plt.scatter(dob_array, y_values, color='blue', label='')  # 
plt.plot(
    dob_array,
    slope * dob_array + intercept,
    color='red',
    label=f': y={slope:.2f}x+{intercept:.2f}'  #  
)

#  
plt.title('    ')
plt.xlabel('')
plt.ylabel('   ')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#  x- ()
dob_array = np.array([1990, 1991, 1992, 1993, 1994, 1995, 1996, 1997, 1998, 1999, 2000])

# Y-   (   )
y_values = np.arange(len(dob_array))

#  
sns.set(style="whitegrid")

#  
plt.figure(figsize=(10, 6))
sns.scatterplot(x=dob_array, y=y_values, color='blue', label='')  # 
sns.regplot(x=dob_array, y=y_values, ci=None, color='red', label=' ')  #  

#  
plt.title('    ')
plt.xlabel('')
plt.ylabel('   ')
plt.legend()
plt.grid(True)
plt.show()
```
